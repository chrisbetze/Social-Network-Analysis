```python
# Ορίζουμε τον αριθμό κόμβων n=1x0 που θα χρησιμοποιήσουμε για την κατασκευή των συνθετικών τοπολογιών.
# Όπου x αντικαταστήσαμε το τελευταίο ψηφίο του αριθμού μητρώου μας. (Α.Μ.: 03116067)

n=170
```

```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Α) Δημιουργία και οπτικοποίηση σύνθετων τύπων δικτύου

Κατασκευή και οπτικοποίηση της τοπολογίας REG για βαθμό πλέγματος $d = 4$

```python
d = 4

# Για την κατασκευή του REG χρησιμοποιήσαμε την watts_strogatz_graph από το networkx. Για Regular επιλέγουμε gp=0. 
REG= nx.watts_strogatz_graph(n, d, 0)

# Για την οπτικοποίηση, χρησιμοποιήσαμε την draw_circular από το networkx
plt.figure(figsize=(8,8))
nx.draw_circular(REG,node_size=50)
```

Κατασκευή και οπτικοποίηση της τοπολογίας RG(ER) για αριθμό συνδέσεων $M = 750$

```python
M = 750

# Για την κατασκευή του RGER χρησιμοποιήσαμε την gnm_random_graph από το networkx
RGER= nx.gnm_random_graph(n,M)

# Για την οπτικοποίηση, χρησιμοποιήσαμε την draw_circular από το networkx
plt.figure(figsize=(8,8))
nx.draw_circular(RGER,node_size=50)
```

Κατασκευή και οπτικοποίηση της τοπολογίας RGG σε περιοχή $L\times L = 1$ (unit cube) και ακτίνα σύνδεσης $R=0.25$

```python
R = 0.25

# Για την κατασκευή του RGG, χρησιμοποιήσαμε την random_geometric_graph από το networkx
RGG= nx.random_geometric_graph(n,R,2)

# Για την οπτικοποίηση, χρησιμοποιήσαμε draw_network_edges, draw_network_nodes. 
# Για τη θέση των κόμβων στην τετράγωνη περιοχή, χρησιμοποιήσαμε την get_node_attributes από το networkx. 
pos= nx.get_node_attributes(RGG,'pos')
plt.figure(figsize=(10,10))
nx.draw_networkx_nodes(RGG,pos,node_size=50)
nx.draw_networkx_edges(RGG,pos)
```

Κατασκευή και οπτικοποίηση της τοπολογίας SF για βαθμό αρχικού πλέγματος $d=4$

```python
d = 4

# Για την κατασκευή του SF, χρησιμοποιήσαμε την barabasi_albert_graph από το networkx.
SF = nx.barabasi_albert_graph(n,d)

# Για την οπτικοποίηση, χρησιμοποιήσαμε την draw_circular από το networkx.
plt.figure(figsize=(8,8))
nx.draw_circular(SF,node_size=50)
```

Κατασκευή και οπτικοποίηση της τοπολογίας SW για βαθμό αρχικού πλέγματος $d=4$ και πιθανότητας ανασύνδεσης $p=0.3$

```python
d = 4
p = 0.3

# Για την κατασκευή του SW, χρησιμοποιήσαμε την watts_strogatz_graph από το networkx.
SW = nx.watts_strogatz_graph(n,d,p)

# Για την οπτικοποίηση, χρησιμοποιήσαμε την draw_circular από το networkx.
plt.figure(figsize=(8,8))
nx.draw_circular(SW,node_size=50)
```

### B) Μελέτη βαθμού κόμβων

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS)

##### 1)  Να υπολογιστεί ο βαθμός κάθε κόμβου

```python
# Χρησιμοποιήσαμε την degree του networkx 

# Συνάρτηση η οποία επιστρέφει μια λίστα με τον βαθμό κάθε κόμβου
def find_degree(G): 
  res = nx.degree(G)
  return [i[1] for i in res]

#Εκτύπωση των βαθμών για όλες τις τοπολογίες
REG_degree= find_degree(REG)
print("Degree of each node")
print("")
print("REG: " + str(REG_degree))
RGER_degree=find_degree(RGER)
print("RGER: " + str(RGER_degree))
RGG_degree=find_degree(RGG)
print("RGG: " + str(RGG_degree))
SF_degree=find_degree(SF)
print("SF: " + str(SF_degree))
SW_degree=find_degree(SW)
print("SW: " + str(SW_degree))
```

##### 2)  Να αναπαρασταθεί (plot) η κατατανομή βαθμών κόμβων του δικτύου (δηλαδή, για κάθε βαθμό πόσοι κόμβοι έχουν αυτό το βαθμό) καθώς και η συγκεντρωτική κατανομή βαθμού κόμβου (δηλαδή για κάθε βαθμό πόσοι κόμβοι έχουν το πολύ αυτό το βαθμο)

```python
import collections

# Συνάρτηση που επιστρέφει μετρικές για την κατανομή και την συγκεντρωτική κατανομή
# Θα χρησιμοποιηθεί παρακάτω και για άλλες κατανομές εκτός της degree

def plot_distribution(G_degree):
  degree_count = collections.Counter(G_degree)
  degree_count_sorted = sorted(degree_count.items(), key=lambda pair: pair[0], reverse=False)
  deg, cnt = zip(*degree_count_sorted)
  cnt_cumsum = np.cumsum(cnt)
  cnt_cumsum = cnt_cumsum/max(cnt_cumsum)
  return (deg, cnt, cnt_cumsum)

#Διαγράμματα Κατανομής
fig, axs = plt.subplots(2,3,figsize=(18,8))

(a1,b1,c1) = plot_distribution(REG_degree)
axs[0,0].bar(a1, b1, width=0.5)
axs[0,0].set_xlim(0,8)
axs[0,0].set_title("REG")
axs[0,0].set_xlabel('Degree')
axs[0,0].set_ylabel('Number of Nodes')

(a2,b2,c2) = plot_distribution(RGER_degree)
axs[0,1].bar(a2, b2, width=0.5)
axs[0,1].set_title("RGER")
axs[0,1].set_xlabel('Degree')
axs[0,1].set_ylabel('Number of Nodes')

(a3,b3,c3) = plot_distribution(RGG_degree)
axs[0,2].bar(a3, b3, width=0.5)
axs[0,2].set_title("RGG")
axs[0,2].set_xlabel('Degree')
axs[0,2].set_ylabel('Number of Nodes')

(a4,b4,c4) = plot_distribution(SF_degree)
axs[1,0].bar(a4, b4, width=0.5)
axs[1,0].set_title("SF")
axs[1,0].set_xlabel('Degree')
axs[1,0].set_ylabel('Number of Nodes')

(a5,b5,c5) = plot_distribution(SW_degree)
axs[1,1].bar(a5, b5, width=0.5)
axs[1,1].set_title("SW")
axs[1,1].set_xlabel('Degree')
axs[1,1].set_ylabel('Number of Nodes')

axs[1,2].axis('off')
fig.tight_layout()

# Διαγράμματα Συγκεντρωτικής Κατανομής
# Χρησιμοποιήσαμε την plot(X,Y) του matplotlib με drawstyle την "steps-post"

fig, axs = plt.subplots(2,3,figsize=(18,8))

axs[0,0].scatter(a1, c1)
axs[0,0].set_title("REG")
axs[0,0].set_xlabel('Cumulative Degree')
axs[0,0].set_ylabel('Percentage')

axs[0,1].plot(a2, c2, drawstyle='steps-post')
axs[0,1].set_title("RGER")
axs[0,1].set_xlabel('Cumulative Degree')
axs[0,1].set_ylabel('Percentage')

axs[0,2].plot(a3, c3, drawstyle='steps-post')
axs[0,2].set_title("RGG")
axs[0,2].set_xlabel('Cumulative Degree')
axs[0,2].set_ylabel('Percentage')

axs[1,0].plot(a4, c4, drawstyle='steps-post')
axs[1,0].set_title("SF")
axs[1,0].set_xlabel('Cumulative Degree')
axs[1,0].set_ylabel('Percentage')

axs[1,1].plot(a5, c5, drawstyle='steps-post')
axs[1,1].set_title("SW")
axs[1,1].set_xlabel('Cumulative Degree')
axs[1,1].set_ylabel('Percentage')

axs[1,2].axis('off')
fig.tight_layout()
```

<b>Σχολιάζουμε εδώ τα αποτελέσματα των ερωτημάτων Β1, Β2, Β3.</b>
<br> 
*   REG: Όπως αναμέναμε και από τη θεωρία για τα Regular γραφήματα ο βαθμός κόμβου είναι σταθερός και ίσος με την παράμετρο d που του ορίσαμε στην αρχή. Η κατατανομή βαθμών κόμβων του δικτύου είναι η συνάρτηση dirac και η συγκεντρωτική κατανομή είναι προγανώς ένα μόνο σημείο. Γι αυτό βλέπουμε και η μέση τιμή να ισούται με το d και η διασπορά να είναι μηδέν.

*   RGER: Από θεωρία γνωρίζουμε ότι για τυχαίους γράφους με σχετικά μεγάλο $n$ και σταθρό $np$, η κατανομή βαθμού κόμβων είναι κατανομή Poisson με $λ=np$. Αυτό γίνεται ξεκάθαρα εμφανές από το διάγραμμα της προσομοίωσή μας. Επιπλέον γνωρίζουμε ότι με πιθανότητα $p$, η μέση τιμή βαθμού κόμβου είναι περίπου $mean=p(n-1)$, το ίδιο και η διασπορά. Αν ανάξουμε το δικό μας $M=750$ σε $p$, έχουμε ότι $p=\frac{2M}{n(n-1)}=\frac{1500}{28730}=0.05221$ και άρα $mean=0.05221*169=8.8235$ όσο δηλαδή βρήκαμε κι εμείς παραπάνω με την προσομοίωση. Λογικά, αν αυξάναμε παραπάνω το n και μειώναμε κατάλληλα το p ώστε το np=σταθ, τότε θα διαπιστώναμε ότι και το variance θα πλησίαζε την ίδια τιμή περισσότερο.

*   RGG: Για Random Geometric Graphs οι κόμβοι είναι διατεταγμένοι τυχαία στο τετράγωνο 1x1 και οι συνδέσεις γίνονται σε ακτίνα R=0.25 γύρω από τον κάθε κόμβο, γι αυτό δημιουργούνται σαν ομάδες κόμβων γεωγραφικά, κατανεμημένα στο δίκτυο. Η κατανομή κόμβων είναι μια καμπάνα ομοιόμορφη (Uniform) χωρίς εμφανή μέγιστη τιμή. Είναι δηλαδή πιο κατανεμημένοι οι βαθμοί των κόμβων κι αυτό φαίνεται και από την συγκετρωτική κατανομή. Επιπλέον, από θεωρία γνωρίζουμε ότι ο αναμενόμενος βαθμός κόμβου είναι $mean=\frac{πr^2}{L^2}n=33.18$, που είναι κοντά στην μέση τιμή που βρήκαμε εμείς. Η διασπορά προφανώς και είναι αρκετά μεγάλη διότι όπως είπαμε υπάρχει αυτή η ομοιομορφία σε μεγάλο εύρος τιμών.

*   SF: Για τα Scale Free δίκτυα, όπως βλέπουμε από τα παραπάνω διαγράμματα αλλά και όπως ξέρουμε και από θεωρία, μικρό ποσοστό των κόμβων έχουν μεγάλο βαθμό κόμβου. Δηλαδή, η πλειοψηφία των κόμβων έχει μικρές τιμές βαθμού, έχουμε Power law degree distribution. Αυτό γίνεται φανερό κι από την απεικόνιση του δικτύου που φαίνεται στο ερώτημα Α. Η μέση τιμή είναι λίγο μεγαλύτερη από τον βαθμό αρχικού πλέγματος (d=4), ενώ η διασπορά είναι αρκετά μεγάλη (τετραπλάσια από την μέση τιμή).

*   SW: Τα Small World δίκτυα είναι ανάμεσα στα Regular και στα Random, καθώς αυξάνουμε την τιμή της πιθανότητας ανασύδεσης από 0 μέχρι 1 (συνήθως όχι μεγαλύτερες από 0.5). Έχουμε όπως βλέπουμε από τα διαγράμματα κατανομή heavy-tailed, σχετικά ομογενή τοπολογία και όπως ξέρουμε και από τη θεωρία οι περισσότεροι κόμβοι έχουν περίπου τον ίδιο αριθμό γειτόνων. Αυτό διαπυστώνεται και από την διασπορά, η οποία είναι σχετικά μικρή (κάτω από 1) και η μέση τιμή πολύ κοντά στον βαθμό του αρχικού πλέγματος (d=4). Δηλαδή, όσο αυξάνει το gp τόσο πλησιάζει την Random (ανοίγει-πλαταίνει η κατανομή), ενώ όσο μικραίνει η πιθανότητα τόσο στενεύει το διάγραμμα ώστε να φτάσει στη dirac (Regular).



### Γ) Δίκτυα με βάρη

Για ένα δίκτυο με βάρη στις συνδέσεις του, όμοια με τον πίνακα γειτνίασης, ορίζεται ο πίνακας βαρών, $W=[w_{ij}]$, όπου με $w_{ij}>0$ συμβολίζεται το βάρος στην ακμή που ενώνει τους κόμβους $i,j$. Για μη-κατευθυνόμενους γράφους ο πίνακας $W$ είναι συμμετρικός.

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να δημιουργηθεί τυχαίος πίνακας βαρών. Το εύρος των βαρών να κυμαίνεται στο διάστημα $[1,10]$.

```python
a=1
b=10
# Για την προσθήκη βαρών στις ακμές, χρησιμοποήσαμε την add_edge του networkx

# Συνάρτηση που προσθέτει τυχαία βάρη στον γράφο G στο διάστημα [1,10]
def add_rand_weights(G): 
  for i in list(G.edges):
    G.add_edge(*i,weight=np.random.rand()*(b-a) + a)

# Εκτύπωση ακμών με τα βάρη τους για κάθε τοπολογία
for i in [REG, RGER, RGG, SF, SW]:
  add_rand_weights(i)
  print(list(i.edges(data='weight')))
```

##### i) Να υπολογιστεί η δύναμη του κάθε κόμβου

```python
# Μαε δόθηκε η συνάρτηση strength που έχει ως όρισμα την τοπολογία που παράξατε στο ερώτημα Α με τα βάρη στις ακμές:
# python 3 implementation

def strength(topology):
    weights_list=[]
    for i in range(0,len(topology.adj)):
        a=[]
        for j in range (0,len(topology.adj[i].items())):
            d=list(topology.adj[i].values())[j]
            c=list(d.values())[0]
            a.append(c)
        weights_list.append(a)
    node_weights = []
    for node in weights_list:
        node_weights.append(sum(node))
    return node_weights

# Υπολογισμός δύναμης για κάθε κόμβο κάθε τοπολογίας
REG_strength = strength(REG)
RGER_strength = strength(RGER)
RGG_strength = strength(RGG)
SF_strength = strength(SF)
SW_strength = strength(SW)

# Εκτύπωση δύναμης κάθε κόμβου για κάθε τοπολογία
print(REG_strength)
print(RGER_strength)
print(RGG_strength)
print(SF_strength)
print(SW_strength)
```

### Δ) Υπολογισμός μέσου μήκους μονοπατιού

```python
# Χρησιμοποιούμε την all_pairs_shortest_path_length του networkx 

# Η συνάρτηση επιστρέφει μια λίστα από λίστες των n=170 κόμβων
def compute_path_lengths(G):
  res=[]
  length = dict(nx.all_pairs_shortest_path_length(G))
  for i in length.keys():
    res.append(list(length[i].values()))
  return res


REG_path_lengths = compute_path_lengths(REG)
RGER_path_lengths = compute_path_lengths(RGER)
RGG_path_lengths = compute_path_lengths(RGG)
SF_path_lengths = compute_path_lengths(SF)
SW_path_lengths = compute_path_lengths(SW)

# Εκτύπωση του mean path και του variance of path για κάθε τοπολογία μέσω του np.mean και np.var
print("Mean path length REG: %f" %np.mean(REG_path_lengths))
print("Variance of path length REG: %f" %np.var(REG_path_lengths))

print("Mean path length RGER: %f" %np.mean(RGER_path_lengths))
print("Variance of path length RGER: %f" %np.var(RGER_path_lengths))

print("Mean path length RGG: %f" %np.mean(RGG_path_lengths))
print("Variance of path length RGG: %f" %np.var(RGG_path_lengths))

print("Mean path length SF: %f" %np.mean(SF_path_lengths))
print("Variance of path length SF: %f" %np.var(SF_path_lengths))

print("Mean path length SW: %f" %np.mean(SW_path_lengths))
print("Variance of path length SW: %f" %np.var(SW_path_lengths))
```

##### 2) Υπολογισμός ΣΟ σε μεγαλύτερες συνθετικές τοπολογίες

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να υπολογιστεί ο τοπικός ΣΟ του κάθε κόμβου, να απεικονιστεί η συγκετρωτική κατανομή ΣΟ για όλους τους κόμβους και να σχολιαστούν τα αποτελέσματα.

```python
# Χρησιμοποιήσαμε την clustering του networkx για τον υπολογισμό του ΣΟ

# Συνάρτηση που επιστρέφει μια λίστα με τα τοπικά ΣΟ για κάθε κόμβο
def calculate_clust_coef(G):
  return list(nx.clustering(G).values())

REG_clust_coef = calculate_clust_coef(REG)
RGER_clust_coef = calculate_clust_coef(RGER)
RGG_clust_coef = calculate_clust_coef(RGG)
SF_clust_coef = calculate_clust_coef(SF)
SW_clust_coef = calculate_clust_coef(SW)

# Τυπώνουμε τον ΣΟ για κάθε κόμβο
print("Συντελεστής Ομαδοποίησης για κάθε κόμβο")
print("")
print("REG: "+str(REG_clust_coef))
print("RGER: "+str(RGER_clust_coef))
print("RGG: "+str(RGG_clust_coef))
print("SF: "+str(SF_clust_coef))
print("SW: "+str(SW_clust_coef))

# Η συνάρτηση αυτή είναι ιδια με την degree distribution που χρησιμοποιήσαμε στο Β ερώτημα με την μόνη
# διαφορά ότι τώρα θα πάρει σαν είσοδο τα G_clust_coef αντί για τα G_degree

def plot_distribution(G_clust_coef):
  degree_count = collections.Counter(G_clust_coef)
  degree_count_sorted = sorted(degree_count.items(), key=lambda pair: pair[0], reverse=False)
  deg, cnt = zip(*degree_count_sorted)
  cnt_cumsum = np.cumsum(cnt)
  cnt_cumsum = cnt_cumsum/max(cnt_cumsum)
  return (deg, cnt, cnt_cumsum)

# Plots για την συγκεντρωτική κατανομή του ΣΟ
fig, axs = plt.subplots(2,3,figsize=(18,8))

(a,b,c) = plot_distribution(REG_clust_coef)
axs[0,0].scatter(a, c)
axs[0,0].set_title("REG")
axs[0,0].set_ylabel("Percentage")
axs[0,0].set_xlabel("Cumulative C.C.")

(a,b,c) = plot_distribution(RGER_clust_coef)
axs[0,1].plot(a, c, drawstyle='steps-post')
axs[0,1].set_title("RGER")
axs[0,1].set_ylabel("Percentage")
axs[0,1].set_xlabel("Cumulative C.C.")

(a,b,c) = plot_distribution(RGG_clust_coef)
axs[0,2].plot(a, c, drawstyle='steps-post')
axs[0,2].set_title("RGG")
axs[0,2].set_ylabel("Percentage")
axs[0,2].set_xlabel("Cumulative C.C.")

(a,b,c) = plot_distribution(SF_clust_coef)
axs[1,0].plot(a, c, drawstyle='steps-post')
axs[1,0].set_title("SF")
axs[1,0].set_ylabel("Percentage")
axs[1,0].set_xlabel("Cumulative C.C.")

(a,b,c) = plot_distribution(SW_clust_coef)
axs[1,1].plot(a, c, drawstyle='steps-post')
axs[1,1].set_title("SW")
axs[1,1].set_ylabel("Percentage")
axs[1,1].set_xlabel("Cumulative C.C.")

axs[1,2].axis('off')
fig.tight_layout()
```

```python
# Εκτυπώνουμε τον μέσο ΣΟ κάθε τοπολογίας
print("Mean clustering coefficient REG: "+str(np.mean(REG_clust_coef)))
print("Mean clustering coefficient RGER: "+str(np.mean(RGER_clust_coef)))
print("Mean clustering coefficient RGG: "+str(np.mean(RGG_clust_coef)))
print("Mean clustering coefficient SF: "+str(np.mean(SF_clust_coef)))
print("Mean clustering coefficient SW: "+str(np.mean(SW_clust_coef)))
```

##### 2) Υπολογισμός κεντρικότητας σε μεγαλύτερες συνθετικές τοπολογίες

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να υπολογιστεί η τοπική κεντρικότητα του κάθε κόμβου, να απεικονιστεί η συγκεντρωτική κατανομή κεντρικότητας για όλους τους κόμβους και να σχολιαστούν τα αποτελέσματα για όλες τις παρακάτω μετρικές κεντρικότητας:

- degree centrality
- closeness centrality

```python
# Συναρτήσεις που επιστρέφουν τις κεντρικότητες ανά κόμβο για έναν γράφο G

def calculate_degree_centrality(G):
  degree_centrality = list(nx.degree_centrality(G).values())
  return degree_centrality

def calculate_closeness_centrality(G):
  closeness_centrality = list(nx.closeness_centrality(G).values())
  return closeness_centrality

def calculate_betweenness_centrality(G):
  betweenness_centrality = list(nx.betweenness_centrality(G).values())
  return betweenness_centrality

def calculate_eigenvector_centrality(G):
  eigenvector_centrality = list(nx.eigenvector_centrality(G).values())
  return eigenvector_centrality

REG_degree_centrality = calculate_degree_centrality(REG)
REG_closeness_centrality = calculate_closeness_centrality(REG)
REG_betweenness_centrality = calculate_betweenness_centrality(REG)
REG_eigenvector_centrality = calculate_eigenvector_centrality(REG)

RGER_degree_centrality = calculate_degree_centrality(RGER)
RGER_closeness_centrality = calculate_closeness_centrality(RGER)
RGER_betweenness_centrality = calculate_betweenness_centrality(RGER)
RGER_eigenvector_centrality = calculate_eigenvector_centrality(RGER)

RGG_degree_centrality = calculate_degree_centrality(RGG)
RGG_closeness_centrality = calculate_closeness_centrality(RGG)
RGG_betweenness_centrality = calculate_betweenness_centrality(RGG)
RGG_eigenvector_centrality = calculate_eigenvector_centrality(RGG)

SF_degree_centrality = calculate_degree_centrality(SF)
SF_closeness_centrality = calculate_closeness_centrality(SF)
SF_betweenness_centrality = calculate_betweenness_centrality(SF)
SF_eigenvector_centrality = calculate_eigenvector_centrality(SF)

SW_degree_centrality = calculate_degree_centrality(SW)
SW_closeness_centrality = calculate_closeness_centrality(SW)
SW_betweenness_centrality = calculate_betweenness_centrality(SW)
SW_eigenvector_centrality = calculate_eigenvector_centrality(SW)



# Για πιο εύκολο σχολιασμό, παράγουμε 1 figure για όλες τις τοπολογίες ανά κεντρικότητα
def plot_centrality(a,b,c,d,e,f):
  plt.figure(figsize=(10,8))
  plt.title(f)
  (x,y,z) = plot_distribution(a)
  plt.scatter(x, z, label="REG")
  (x,y,z) = plot_distribution(b)
  plt.plot(x, z, drawstyle='steps-post', label="RGER")
  (x,y,z) = plot_distribution(c)
  plt.plot(x, z, drawstyle='steps-post', label="RGG")
  (x,y,z) = plot_distribution(d)
  plt.plot(x, z, drawstyle='steps-post', label="SF")
  (x,y,z) = plot_distribution(e)
  plt.plot(x, z, drawstyle='steps-post', label="SW")
  plt.legend()
  plt.show()

plot_centrality(REG_degree_centrality, RGER_degree_centrality, RGG_degree_centrality, SF_degree_centrality, SW_degree_centrality, "Degree Centrality")
plot_centrality(REG_closeness_centrality, RGER_closeness_centrality, RGG_closeness_centrality, SF_closeness_centrality, SW_closeness_centrality, "Closeness Centrality")
plot_centrality(REG_betweenness_centrality, RGER_betweenness_centrality, RGG_betweenness_centrality, SF_betweenness_centrality, SW_betweenness_centrality, "Betweenness Centrality")
plot_centrality(REG_eigenvector_centrality, RGER_eigenvector_centrality, RGG_eigenvector_centrality, SF_eigenvector_centrality, SW_eigenvector_centrality, "Eigenvector Centrality")
```

*   <u>Degree Centrality</u>: Τα αποτελέσματα για το μέσο Degree Centrality εξαρτώνται σημαντικά από τον μέσο βαθμό κόμβων που υπολογίσαμε στο Β ερώτημα. O RGG είναι αυτός με τη μεγαλύτερη κατά μέσο όρο τιμή, αυξάνεται ομοιόμορφα κι αυτό φαίνεται και από το διάγραμμα. Ήταν αυτός με τον μεγαλύτερο βαθμό κόμβων. O SW έχει παρόμοια μέση τιμή με τον REG διότι έχουν παρόμοιο μέσο βαθμό κόμβων και το διάγραμμα συγκντρωτικής κατανομής είναι το πιο απότομο από τα υπόλοιπα. Ο RGER έχει μερικά παρόμοια χαρακτηριστικά με τον RGG, δηλαδή μια ομοιομορφία στο διάγραμμα, αλλά πολύ μικρότερη μέση τιμή προφανώς. Ο SF έχει οριμσένους κόμβους με πολλούς γείτονες (το διάγραμμα απλώνεται και σε πιο μεγάλες τιμές), παρόλο αυτά είναι πολύ λίγοι για να του αυξήσουν την μέση τιμή.

*   <u>Closeness Centrality</u>: Εξαρτάται από την απόσταση των κόμβων από όλους τους άλλους κόμβους, επομένως η μέση closeness centrality εξαρτάται σημαντικά από το μέσο μήκος μονοπατιού που βρήκαμε στο Δ ερώτημα. Επειδή εμείς βρίσκουμε την κανονικοποιημένη τιμή (πολλαπλασιάζουμε με το n-1), χαμηλή τιμή centrality σημαίνει σχετικά μεγάλο μέσο μήκος μονοπατιού. Όπως και πριν προφανώς το REG έχει σταθερή τιμή αφού έχει σταθερούς γείτονες και άρα σταθερές αποστάσεις. Ο SF έχει την υψηλότερη μέση τιμή (αυτό διακρίνεται και από το διάγραμμα) διότι όπως γνωρίζουμε και από τη θεωρία είναι αυτός με τα πιο μικρά μονοπάτια. Ο RGER και ο RGG έχουν κι εδώ παρόμοια χαρακτηριστικά και στο διάγραμμα και στη μέση τιμή, ενώ ο SW είναι ο πιο χαμηλός (εκτός του REG), γεγονός που έρχεται λίγο σε αντίθεση με τη θεωρία, πράγμα που μπορεί να οφείλεται στην χαμηλή τιμή της πιθανότητας p.

*   <u>Betweenness Centrality</u>: Εξαρτάται από το πόσο συμμετέχει ο κάθε κόμβος στα ελάχιστα μονοπάτια των άλλων κόμβων. Επειδή πάλι μελετάμε την κανονικοποιημένη τιμή, υψηλό betweenness centrality σημαίνει μικρή συμμετοχή στην κεντρικότητα. Ο REG είναι πάλι ο "χειρότερος" με σταθερή τιμή κεντρικότητας. Από το διάγραμμα φαίνεται ότι ο RGER, o RGG και ο SW έχουν μια παρόμοια καμπύλη, με τον SW να έχει την μαγαλύτερη μέση τιμή. Ο SF απλώνεται πάλι σε μεγάλο εύρος τιμών, αλλά παρόλο αυτά η μέση τιμή του είναι η πιο χαμηλή, άρα έχει την καλύτερη μέση κεντρικότητα.

*   <u>Eigenvector Centrality</u>: Σχετικά σκορ εκχωρούνται σε όλους τους κόμβους του δικτύου με βάση την ιδέα ότι οι συνδέσεις με κόμβους υψηλού σκορ συμβάλλουν περισσότερο στο σκορ του εν λόγω κόμβου, από τις ίσες συνδέσεις με τους κόμβους χαμηλού σκορ. Η υψηλή τιμή του eigenvector σημαίνει ότι ένας κόμβος συνδέεται με πολλούς κόμβους που οι ίδιοι έχουν υψηλά σκορ. Επειδή εμείς βρίσκουμε την κανονικοποιημένη τιμή, ισχύει το ανάποδο. Δηλαδή, ο SF έχει την χαμηλότερη μέση τιμή, άρα την "καλύτερη" eigenvector centrality, συγκριτικά με τον REG, ο οποίος έχει σταθερή τιμή, την "χειρότερη" από την μέση τιμή των υπολοίπων. Οι γραφικές παραστάσεις παρουσιάζουν εδώ την μεγαλύτερη ομοιομορφία μεταξύ τους, συγκριτικά με τις προηγούμενες κεντρικότητες που μελετήσαμε.

```python
# Για τον έλεγχο της συνεκτικότητας, χρησιμοποίησαμε την is_connected του networkx
# Για τη γραφική αναπαράσταση των αποτελεσμάτων, χρησιμοποίησαμε την bar του matplotlib

# Έλεγχος συνεκτικότητας για REG

# Δημιουργούμε 100 τοπολογίες για κάθε δυνατό συνδιασμό και βλέπουμε πόσες από αυτές είναι συνεκτικές
# Άρα στην ουσία έχουμε ένα είδος ποσοστού

for n in [100,200]:
  reg=[]
  for d in range (2,12,2):
    connect = 0
    for i in range(100):
      if(nx.is_connected(nx.watts_strogatz_graph(100, d, 0))):
        connect=connect+1
    reg.append(connect) 
    
  plt.bar(list(range(2,12,2)),reg, width=0.5, color='b')
  plt.title("REG n = %d" %n)
  plt.ylabel("Number of connected graphs")
  plt.xlabel("Degree")
  plt.show()
```

Ενώ, όταν είναι από 400 και πάνω οι συνδέσεις, η πιθανότητα αγγίζει το 100%. Αντίστοιχα, για $n$=200 κόμβους και συνδέσεις κάτω από τις 400, η πιθανότητα είναι 0%, ενώ για συνδέσεις $Μ$ μεγαλύτερες από 800, η πιθανότητα είναι 100%.


```python
# Έλεγχος συνεκτικότητας για RG

# Δημιουργούμε 100 τοπολογίες για κάθε δυνατό συνδιασμό και βλέπουμε πόσες από αυτές είναι συνεκτικές
# Άρα στην ουσία έχουμε ένα είδος ποσοστού

for n in [100,200]:
  rg=[]
  for p in np.arange(0.1, 1.0, 0.1):
    connect = 0
    for i in range(100):
      if(nx.is_connected(nx.gnp_random_graph(n,p))):
        connect=connect+1
    rg.append(connect) 
    
  plt.bar(list(np.arange(0.1, 1.0, 0.1)), rg, width=0.05, color='b')
  plt.title("RG n = %d" %n)
  plt.ylabel("Number of connected graphs")
  plt.xlabel("Probability")
  plt.show()
```

<u>Σχολιάστε τα αποτελέσματα συνεκτικότητας για το **RG Gilbert**</u>:
<br>
Όταν έχουμε έναν $G(n,p)$ graph, στην ουσία έχουμε αρχικά $n$ απομονωμένους κόμβους και προσθέτουμε διαδοχικά ακμές μεταξύ αυτών, ανεξάρτητα η μία από την άλλη με πιθανότητα $p$. Ο αναμενόμενος αριθμός συνδέσεων για το μοντέλο RG Gilbert είναι ${n\choose2}p$. Με βάση τα παραπάνω διαγράμματα, για $n$=100 και $n$=200, οποιαδήποτε πιθανότητα και να επιλεχθεί στο διάστημα [0.1, 0.9], το δίκτυο που θα προκύψει θα είναι συνεκτικό. To $p$=0.1 και $n$=100 αντιστοιχεί με βάση τη θεωρία και το παραπάνω τύπο περίπου σε $M$=495. Επομένως, η πρώτη στύλη του "RG n=100" διαγράμματος αντιστοιχεί περίπου με την $5^η$ στύλη ($Μ$=500) του "RGER n=100" διαγράμματος η οποία είδαμε προηγουμένως ότι είχε πιθανότητα 100%. Αντίσοιχα, το $p$=0.1 και $n$=200 αντιστοιχεί περίπου σε $M$=1990. Επομένως, η πρώτη στύλη του "RG n=200" διαγράμματος αντιστοιχεί περίπου με στύλη του "RGER n=200" πολύ πιο δεξιά του 800, η οποία προφανώς και δίνει πιθανότητα σχεδόν 100%. Άρα, όσο αυξάνεται το $n$ και η πιθανότητα μένει στην ίδια τάξη μεγέθους, τόσο πιο πιθανό είναι να είναι συνεκτικός ο γράφος. Επομένως, πρέπει να μειωθεί κι άλλο η πιθανότητα για να προκύψει μη συνεκτικός γράφος.

```python
# Έλεγχος συνεκτικότητας για RGG

# Δημιουργούμε 100 τοπολογίες για κάθε δυνατό συνδιασμό και βλέπουμε πόσες από αυτές είναι συνεκτικές
# Άρα στην ουσία έχουμε ένα είδος ποσοστού
for n in [100,200]:
  rgg=[]
  for R in np.arange(0.025, 0.275, 0.025):
    connect = 0
    for i in range(100):
      if(nx.is_connected(nx.random_geometric_graph(n,R,2))):
        connect=connect+1
    rgg.append(connect) 
    
  plt.bar(list(np.arange(0.025, 0.275, 0.025)),rgg, width=0.016, color='b')
  plt.title("RGG n = %d" %n)
  plt.ylabel("Number of connected graphs")
  plt.xlabel("Radius")
  plt.show()
```

<u>Σχολιάστε τα αποτελέσματα συνεκτικότητας για το **RGG**</u>:
<br>
Παρατηρούμε ότι για μικρή τιμή της ακτίνας λογικό να μην έχουμε μεγάλο ποσοστό συνεκτικών γράφων, διότι είναι πιο δύσκολο να βρεθεί ο γείτονας σε μικρότερη ακτίνα. Από την άλλη, όταν αυξάνεται ο αριθμός των κόμβων $n$, κατανέμονται πιο πολύ κόμβοι στο τετράγωνο και άρα είναι πιο πιθανό να βρεθεί γείτονας στο ίδιο μήκος ακτίνας. Γι αυτό πχ και παρατηρούμε μια μεγάλη αύξησηση του ποσοστού για την ακτίνα 0.15 από 100 σε 200 κόμβους.

```python
# Έλεγχος συνεκτικότητας για SF

# Δημιουργούμε 100 τοπολογίες για κάθε δυνατό συνδιασμό και βλέπουμε πόσες από αυτές είναι συνεκτικές
# Άρα στην ουσία έχουμε ένα είδος ποσοστού
for n in [100,200]:
  sf=[]
  for d in range (2,12,2):
    connect = 0
    for i in range(100):
      if(nx.is_connected(nx.barabasi_albert_graph(n,d))):
        connect=connect+1
    sf.append(connect) 
    
  plt.bar(list(range(2,12,2)),sf, width=0.5, color='b')
  plt.title("SF n = %d" %n)
  plt.ylabel("Number of connected graphs")
  plt.xlabel("Degree Initial Grid")
  plt.show()
```

<u>Σχολιάστε τα αποτελέσματα συνεκτικότητας για το **SF**</u>:
<br>
Το μοντέλο Barabasi-Albert που χρησιμοποιούμε για την δημιουργία Scale Free δικτύου έχει τον εξής αλγόριθμο. Αρχικά, το δίκτυο ξεκινά με έναν συνεκτικό γράφο από κόμβους και στη συνέχεια διαδοχικά νέοι κόμβοι προστίθενται. Κάθε νέος κόμβος που προστίθενται συνδέεται με $d$ από τους ήδη υπάρχοντες κόμβους με πιθανότητα η οποία είναι preferentially attached με κόμβους που έχουν υψηλό βαθμό. Επομένως, υψηλά συνδεδεμένοι κόμβοι (hubs) έχουν την τάση να αποκτήσουν ακόμη παραπάνω γείτονες, ενώ οι κόμβοι με χαμηλό degree έχουν μικρότερη πιθανότητα να επιλεχθούν ανάμεσα από αυτούς του $d$. Δηλαδή η καινούργοι κόμβοι έχουν μια "προτίμηση" να συνδεθούν με κόμβους υψηλού degree. Καταλυκτικά, αυτό που παρατηρούμε από τα παραπάνω διαγράμματα, αλλά γνωρίζουμε ήδη από τη θεωρία είναι ότι όλα τα Scale Free δίκτυα που προκύπτουν από το μοντέλο Barabasi-Albert είναι 100% συνεκτικά, ανεξάρτητα του $n$ και του $d$.

```python
# Έλεγχος συνεκτικότητας για SW

# Δημιουργούμε 100 τοπολογίες για κάθε δυνατό συνδιασμό και βλέπουμε πόσες από αυτές είναι συνεκτικές
# Άρα στην ουσία έχουμε ένα είδος ποσοστού

for prob in np.arange(0.1,0.8,0.1):
  for n in [100,200]:
    sw=[]
    for d in range (2,12,2):
      connect = 0
      for i in range(100):
        if(nx.is_connected(nx.watts_strogatz_graph(n,d,prob))):
          connect=connect+1
      sw.append(connect) 
    if (n==100):
      color='b'
    else:
      color='r' 
    plt.bar(list(range(2,12,2)),sw, width=0.5, color=color)
    plt.title("SW n = "+str(n)+ " and gp = "+str(prob))
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Degree Initial Grid")
    plt.show()
```

<u>Σχολιάστε τα αποτελέσματα συνεκτικότητας για το **SW**</u>:
<br>
Το μοντέλο Watts-Strogatz που χρησιμοποιόυμε για τη δημιουργία Small World δικτύου αρχικά ξεκινάει με έναν Regular Graph $n$ κόμβων οι οποίοι έχουν όλοι βαθμό $d$. Στη συνέχεια, κάθε ακμή έχει πιθανότητα να ανασυνδεθεί ίση με $g_p$. Προφανώς, όπως φαίνεται και από τα παραπάνω διαγράμματα, όσο πιο μεγάλο είναι το $d$, τόσο πιο δύσκολο είναι να ανασυνδεθούν όλες (και οι $d$) οι συνδέσεις ενός κόμβου για να προκύψει τελικά μη συνεκτικός γράφος. Επίσης, η αύξηση της πιθανότητας ανασύνδεσεις, προκαλεί μια μικρή πτώση της πιθανότητας συνεκτικότητας για $d$=2, αλλά καμία αλλαγή για $d$ από 4 και πάνω. Για d δηλαδή από 4 και πάνω, έχουμε πιθανότητα 100%, ανεξάρτητα της $g_p$ και του $n$. Όσο αναφορά την αύξηση του $n$ από τα 100 στα 200, προκαλεί μια μικρή μείωση της πιθανότητας συνεκτικότητας, αλλά και πάλι μόνο για την περίπτωση που το $d$ είναι 2. Επομένως, ερχόμαστε στο συμπέρασμα ότι ο πιο σημαντικός παράγοντας συνεκτικότητας για τα Small World δίκτυα κατά Watts-Strogatz, είναι ο βαθμός του αρχικού πλέγματος.

- Σε ποιες περιπτώσεις εμφανίζονται φαινόμενα κατωφλίου;

Φαινόμενα κατωφλίου παρατηρούμε στους **Random Graphs**, στον **Random Geometric Graph** και στο **Small World** δίκτυο. Όσο αναφορά το **Scale Free** δίκτυο, η θεωρία αναφέρει ότι έχουμε phase transitions, αλλά εδώ που το δημιουργούμε με το μοντέλο Barabasi-Albert,  όπως αναφέραμε και παραπάνω, το δίκτυο ξεκινά με έναν συνεκτικό γράφο από κόμβους και στη συνέχεια διαδοχικά νέοι κόμβοι προστίθενται. Επομένως, σύμφωνα με αυτό το μοντέλο όλοι οι γράφοι θα είναι συνεκτικοί.



- Ποιες είναι οι κρίσιμες τιμές για κάθε τύπο δικτύου οι οποίες οδηγούν σε μετάβαση φάσης (phase transitions);

1. **RG Gilbert**: Αν το $p<\frac{(1-ε)lnn}{n}$, τότε ο γράφος G(n,p) θα είναι σχεδόν σίγουρα μη συνεκτικός
<br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$Αν το $p>\frac{(1+ε)lnn}{n}$, τότε ο γράφος G(n,p) θα είναι σχεδόν σίγουρα συνεκτικός
<br>
Επομένως, η τιμή $p_t=\frac{lnn}{n}$ είναι κρίσιμη τιμή που οδηγεί σε μετάβαση φάσης. Για τα δικά μας παραδείγματα έχουμε n=100 και n=200, άρα $p_t(100)=0.0460$ και $p_t(200)=0.0265$. Στα διαγράμματά μας δεν έχουμε τόσο χαμηλές τιμές της πιθανότητας, επομένως δεν μπορούμε να διακρίνουμε αυτές τις μεταβάσεις φάσης.
<br><br>
2.   **RGER**: Εαν ανάγουμε τα προηγούμενα του Gilbert στο μοντέλο Erdos-Renyi, τότε θα έχουμε $M_t={n\choose2}p_t={n\choose2}\frac{lnn}{n}=\frac{(n-1)lnn}{2}$. Άρα, για n=100 και n=200 έχουμε $M_t(100)=227.96$ και $M_t(200)=527.18$. Αυτές οι τιμές επαληθεύονται όπως βλέπουμε και από τα διαγράμματα του RGER.
<br><br>
3.   **RGG**: Εαν $r^2n≥c_1l^2lnl$ τότε ο RGG(n,r,l) είναι συνεκτικός a.a.s. όπου $c_1>2-2ε$ και $0≤ε<1$ <br>
$\;\;\;\;\;\;\;\;$Εαν $r^2n≤c_0l^2lnl$ τότε ο RGG(n,r,l) είναι μη συνεκτικός a.a.s. όπου $c_0<\frac{1}{2}-\frac{1}{2}ε$ και $0≤ε<1$ <br>
Επομένως, περίπου η τιμή $r_t=\sqrt{\frac{l^2lnl}{n}}$ είναι η κρίσιμη η οποία οδηγεί σε μετάβαση φάσης. Για τα δικά μας παραδείγματα έχουμε n=100 και n=200, άρα $r_t(100)=0.167$ και $r_t(200)=0.118$. Αυτές οι τιμές επαληθεύονται όπως βλέπουμε και από τα διαγράμματα του RGG.
<br><br>
4.   **SF**: Για το SF με το μοντέλο Barabasi-Albert, όπως αναφέραμε και προηγουμένως, δεν παρατηρούμε κάποια μετάβαση φάσης, αφού το δίκτυο είναι σίγουρα συνεκτικό.
<br><br>
5.   **SW**: Για το δίκτυο Small World όπως βλέπουμε και από τα διαγράμματα και αναλύσαμε προηγουμένως, η μόνη κρίσιμη περιοχή μετάβασης φάσης είναι το $d$=4. Για $d$ από 4 και πάνω, ανεξαρτήτως $n$ και $g_p$, η πιθανότητα συνεκτικότητας είναι 100%. Ενώ για χαμηλότερες τιμές υπάρχει μια απότομη πτώση της πιθανότητας.

- Ποιες μεταβάσεις φάσης είναι απότομες (sharp) και ποιες ομαλές (smooth);

1.   **Sharp**: RG Gilbert, RGER, SW

2.   **Smooth**: RGG

### Ι) Μελέτη της εξελικτικής μετατροπής δικτύου REG σε δίκτυο SW και RGER

Μεταβάλλοντας την παράμετρο $g_p$ της συνάρτησης που κατασκευάζει τοπολογίες SW, από $0$ έως $1$ με βήμα $0.1$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγμόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

```python
n = 170
d = 4

G_pl=[]
G_cc=[]

# Δημιουργία δικτύων και υπολογισμός mean path και mean CC για τιμές του gp στο διάστημα [0,1] με βήμα 0.1
for gp in np.arange(0,1.1,0.1):
  G = nx.watts_strogatz_graph(n, d, gp)
  G_pl.append(np.mean(compute_path_lengths(G)))
  G_cc.append(np.mean(calculate_clust_coef(G)))

# Plots για το μέσο μήκος μονοπατιού και το μέσο ΣΟ καθώς μεταβάλλεται η πιθανότητα gp
fig, axs = plt.subplots(1,2,figsize=(18,6))

axs[0].bar(np.arange(0,1.1,0.1),G_pl, width=0.05)
axs[0].set_title("Average Minimum Path")
axs[0].set_xlabel('Probability')

axs[1].bar(np.arange(0,1.1,0.1),G_cc, width=0.05)
axs[1].set_title("Mean Clustering coefficient")
axs[1].set_xlabel('Probability')
plt.show()

# Εκτύπωση mean path και mean CC για κάθε gp
print("Για gp=0.0: Mean Path= "+str(G_pl[0])+" και CC= "+str(G_cc[0]))
print("Για gp=0.1: Mean Path= "+str(G_pl[1])+" και CC= "+str(G_cc[1]))
print("Για gp=0.2: Mean Path= "+str(G_pl[2])+" και CC= "+str(G_cc[2]))
print("Για gp=0.3: Mean Path= "+str(G_pl[3])+" και CC= "+str(G_cc[3]))
print("Για gp=0.4: Mean Path= "+str(G_pl[4])+" και CC= "+str(G_cc[4]))
print("Για gp=0.5: Mean Path= "+str(G_pl[5])+" και CC= "+str(G_cc[5]))
print("Για gp=1.0: Mean Path= "+str(G_pl[10])+" και CC= "+str(G_cc[10]))
```

Σύμφωνα με το μοντέλο αυτό, τυχαία ανασυνδέεται κάθε κορυφή με πιθανότητα $g_p$. Με βάση τα παραπάνω αποτελέσματα, αλλά και το θεωρητικό υπόβαθρο, παρατηρούμε ότι:

*   Για $g_p=0$: Το παραγόμενο δίκτυο συμπεριφέρεται ως ένας Regular Graph, αφού καμία ακμή δεν ανασυνδέεται. Αυτό γίνεται αντιλυπτό και από τα παραπάνω αποτελέσματα. Το Mean_Path και το Mean_CC έχουν ακριβώς τις ίδιες τιμές που βρήκαμε στα προηγούμενα ερωτήματα για το REG.

*   Για $g_p=1$: Το παραγόμενο δίκτυο συμπεριφέρεται ως ένας Random Graph, αφού όλες οι ακμές θα ανασυνδεθούν τυχαία. Αυτό γίνεται αντιλυπτό και από τα παραπάνω αποτελέσματα. Το Mean_Path και το Mean_CC έχουν σχεδόν ακριβώς τις ίδιες τιμές που βρήκαμε στα προηγούμενα ερωτήματα για το RGER.

*   Για $g_p\simeq 0.1-0.5$: Το παραγόμενο δίκτυο συμπεριφέρεται ως ένα Small World network, αφού οι ακμές θα ανασυνδεθούν τυχαία με πιθανότητα $g_p$. Από τα παραπάνω διαγράμματα παρατηρούμε ότι το mean path είναι σχετικά μικρό για SW δίκτυα χωρίς ιδιαίτερες μεταβολές καθώς αυξάνει η πιθανότητα. Αυτό έρχεται σε συμφωνία με τα προηγούμενα ερωτήματα που είδαμε ότι τα SW δίκτυα έχουν περίπου ίδιο μέσο μήκος μονοπατιού με τα RGER. Αντιθέτως, ο μέσος συντελεστής ομαδοποίησης εξαρτάται αρκετά από την πιθανότητα και μειώνεται όσο αυτή αυξάνεται. Γι αυτό και στο ερώτημα Ε που είχαμε $g_p=0.3$ για το SW, είδαμε συντελεστή ομαδοποίησης λίγο μεγαλύτερo από ότι το RGER.

### Κ) Εγω-κεντρικότητες (Ego-centralities)
Για τις συνθετικές τοπολογίες του πίνακα 1, να υπολογιστεί η ενδιαμεσική εγω-κεντρικότητα (ego-betweenness centrality) του κάθε κόμβου. Στη συνέχεια να καταταχθούν οι κόμβοι με βάση την ενδιαμεσική εγωκεντρικότητα και την ενδιαμεσική κεντρικότητα και να συγκριθούν/σχολιαστούν οι κατατάξεις ως προς την ομοιότητά τους για κάθε μία τοπολογία.  

```python
from numpy.linalg import matrix_power
from scipy.sparse import csr_matrix
from seaborn import pairplot
from tabulate import tabulate

# Για την κατασκευή των εγωδικτύων, χρησιμοποιήσαμε την ego_graph του networkx
# Για τον υπολογισμό της ενδιαμεσικής εγωκεντρικότητας θα χρειαστεί να δημιουργήσουμε τον πίνακα γειτνίασης του εγωδικτύου
# Χρήσιμες συναρτήσεις ήταν οι ακόλουθες: 
#   matrix_power(numpy), adj_matrix(networkx), ones(numpy), multiply(numpy), triu(numpy), nonzero(numpy), csr_matrix.toarray(scipy)

# Συνάρτηση υπολογισμού ενδιαμεσικών εγωκεντρικοτήτων 
def calculate_ego_between_centrality(G):
  result = []
  for node in G:    
    ego = nx.ego_graph(G,node)                        # δημιουργία εγωδικτύου
    A = csr_matrix.toarray(nx.adj_matrix(ego))        # Α o πίνακας γειτνίασης του εγωδικτύου
    A2 = matrix_power(A,2)                            # Α2 = Α^2
    IA = np.ones((len(A),len(A))) - A                 # ΙΑ = 1-Α
    RES = np.multiply(A2,IA)                          # RES =  ο πολλαπλασιασμός Hadamard των A^2 με το (Ι-Α)
    TRI = np.triu(RES,1)                              # TRI = τριγωνικός του RES
    FIN = TRI[np.nonzero(TRI)]                        # FIN = περιέχει τις μη αρνητικές τιμές του τριγωνικού πίνακα
    centrality = 0
    for i in FIN:
      centrality = centrality + 1/i   
    result.append(centrality)
  return result                                       # Επιστρέφεται μια λίστα με όλες τις ενδιαμεσικές εγωκεντρικότητες των κόμβων

# Δημιουργία τοπολογιών και υπολογισμός των δύο ειδών κεντρικοτήτων
REG = nx.watts_strogatz_graph(170,4,0)
REG_ego_between_centrality = calculate_ego_between_centrality(REG)
REG_betweenness_centrality = list(nx.betweenness_centrality(REG).values())

RGER= nx.gnm_random_graph(170,750)
RGER_ego_between_centrality = calculate_ego_between_centrality(RGER)
RGER_betweenness_centrality = list(nx.betweenness_centrality(RGER).values())

RGG= nx.random_geometric_graph(170,0.25,2)
RGG_ego_between_centrality = calculate_ego_between_centrality(RGG)
RGG_betweenness_centrality = list(nx.betweenness_centrality(RGG).values())

SF = nx.barabasi_albert_graph(170,4)
SF_ego_between_centrality = calculate_ego_between_centrality(SF)
SF_betweenness_centrality = list(nx.betweenness_centrality(SF).values())

SW = nx.watts_strogatz_graph(170,4,0.3)
SW_ego_between_centrality = calculate_ego_between_centrality(SW)
SW_betweenness_centrality = list(nx.betweenness_centrality(SW).values())

# Κατασκευάσαμε dataframe από το pandas με γραμμές τους κόμβους και στήλες τις τιμές των υπό εξέταση κεντρικοτήτων
# Κατασκευάσαμε στη συνέχεια pairplot από το seaborn για να εξετάσουμε τις συσχετίσεις μεταξύ τους.

# Συνάρτηση για να τυπώνουμε τους κόμβους ανά φθίνουσα σειρά ενδιαμεσικής εγωκεντρικότητα και ενδιαμεσικής κεντρικότητας
# και να εμφανίζει το pairplot.

def plot_sort_nodes_and_pairplot(G, ego, between):
  d = {'ego_between': ego, 'betweenness': between}
  df = pd.DataFrame(data=d)
  g = sns.pairplot(df)
  g.fig.suptitle(G, x=0.58, y=1.05)
  print(G+": Sort by Betweeness")
  print("")
  print(df.sort_values(by=['betweenness'],ascending=False))
  print("")
  print(G+": Sort by Ego Betweeness")
  print("")
  print(df.sort_values(by=['ego_between'],ascending=False))
  print("")

plot_sort_nodes_and_pairplot("REG", REG_ego_between_centrality, REG_betweenness_centrality)
plot_sort_nodes_and_pairplot("RGER", RGER_ego_between_centrality, RGER_betweenness_centrality)
plot_sort_nodes_and_pairplot("RGG", RGG_ego_between_centrality, RGG_betweenness_centrality)
plot_sort_nodes_and_pairplot("SF", SF_ego_between_centrality, SF_betweenness_centrality)
plot_sort_nodes_and_pairplot("SW", SW_ego_between_centrality, SW_betweenness_centrality)
```

**Διαδικασία εύρεσης Ενδιαμεσικής Εγω-Κεντρικότητας για κάθε κόμβο (Ego-Betweenness Centrality):**

*   Βρίσκουμε το εγω-δίκτυο του κόμβου.
*   Δημιουργούμε τον πίνακα γειτνίασης $Α$ του εγω-δικτύου.
*   Από τον πίνακα $1$ (με όλα τα στοιχεία του άσσους) αφαιρούμε τον $Α$ και βρίσκουμε τον $1-Α$.
*   Πολλαπλασιάζουμε κατά Hadamard (στοιχείο προς στοιχείο) τον $A^2$ με τον $(Ι-Α)$, ο οποίος επιστρέφει, για κάθε ζεύγος κόμβων, το πλήθος των ελάχιστων περιπάτων μήκους 2.
*   Απομονώνουμε τον κάτω ή τον πάνω (είναι συμμετρικός) τριγωνικό πίνακα που προκύπτει μετά τον πολλαπλασιασμό.
*   Απομονώνουμε τις μη αρνητικές τιμές, τις αντιστρέφουμε και τις αθροίζουμε.
*   Το αποτέλεσμα που προκύπτει είναι η ενδιαμεσική εγωκεντρικότητα για τον συγκεκριμένο κόμβο.

**Σύγκριση Ενδιαμεσικής Εγω-Κεντρικότητας και Ενδιαμεσικής Κεντρικότητας**

Συνολικά, αυτό που παρατηρούμε από τα sorted dataframes και από τα διαγράμματα είναι ότι για υψηλές τιμές θα υπάρχει αντιστοίχιση μεταξύ των δύο κεντρικοτήτων, φαινόμενο που μειώνεται λιγάκι όσο πέφτουν οι τιμές. Δηλαδή, οι κόμβοι με την μεγαλύτερη Ego-Betweenness Centrality θα είναι πιθανόν και αυτοί με την μεγαλύτερη Betweenness Centrality. Συγκεκριμένα για κάθε τοπολογία έχουμε τα εξής σχόλια:
*   REG: Για Regular Graph προφανώς έχουμε σταθερή τιμή και για τις δύο κεντρικότητες, αφού ο βαθμός κάθε κόμβου είναι σταθερός και ίδιος για όλους τους κόμβους.

*   RGER: Για Random Graph παρατηρούμε μια ομοιομορφία στις δύο κεντρικότητες. Αλλάζουν με παρόμοιο τρόπο, δηλαδή όσο αυξάνεται η μία τόσο αυξάνεται και η άλλη. Το διάγραμμα συσχέτισης δηλαδή είναι νοητά γύρω από μια ευθεία με τα σημεία να είναι σχετικά ομοιόμορφα κατανεμημένα περί της ευθείας. Αυτό επαληθεύεται και από τη σύγκριση των δύο ταξινομημένων dataframes όπου παρατηρούμε ότι οι ίδιοι αριθμοί κόμβων εμφανίζονται και στις δύο περιπτώσεις ταξινόμησης. Υπάρχει δηλαδή σχετική αντιστοιχία στις δύο κεντρικότητες.

*   RGG: Στον Random Geometric Graph έχουμε παρόμοια συμπεριφορά με τον RGER, μόνο που τώρα εδώ τα σημεία είναι λίγο πιο απομακρυσμένα από την ευθεία προς την μεριά του ego_betweenness. Δηλαδή η ενδιαμεσική εγωκεντρικότητα αυξάνεται κατά λίγο πιο γρήγορα συγκριτικά με την ενδιαμεσική κεντρικότητα γεγονός που μπορούμε να το παρατηρήσουμε και στα ταξινομημένα dataframes. (Δεν έχουμε την ίδια αντιστοίχιση με το RGG).

*   SF: Με βάση το διάγραμμα συσχέτισης παρατηρούμε ότι για Scale Free δίκτυο, όλα σχεδόν τα σημεία είναι πάνω σε αυτήν την νοητή ευθεία. Υπάρχει δηλαδή ομοιόμορφη αυξομείωση τιμών, μόνο που τα περισσότερα σημεία εμφανίζονται κοντά στην αρχή των αξόνων. Έχουμε δηλαδή αρκετά χαμηλές τιμές κεντρικοτήτων, γεγονός που το περιμέναμε από τη θωερία και από τα προηγούμενα ερωτήματα. Επιπλέον, αυτό που παρατηρούμε από τα ταξινομημένα dataframes είναι ότι για μεγάλες τιμές κεντικοτήτων υπάρχει αντιστοιχία μεταξύ των δύο ειδών, ενώ όσο πέφτει τόσο χάνεται και αυτή η αντισοιχία.

*   SW: Για το Small World δίκτυο παρατηρούμε από το διάγραμμα ότι τα σημεία μοιάζουν με τον RGG μόνο που εδώ είναι ακόμα περισσότερο μη ομοιόμορφα. Είναι το δίκτυο με τα περισσότερα σημεία αρκετά απομακρυσμένα από την νοητή ευθεία, δηλαδή με το μεγαλύερο εύρος διακύμανσης. Προφανώς αυτό επαληθεύεται και από τα ταξινομημένα dataframes όπου υπάρχουν αρκετές διαφοροποιήσεις.
